```python
single2=perturb.pp.subsample_cells(single,300,'gene.compact')
```

```python
combo2=perturb.pp.subsample_cells(combo,300,'gene.compact')
```

```python
common_genes=list(set(single.var_names).intersection(set(combo.var_names)))
print(len(common_genes))
```

```python
def create_pseudobulk(adata_here,grouping_variable):
        
    #get total counts for each grouping
    df=pd.DataFrame(adata_here.X,
                    index=adata_here.obs_names,
                    columns=adata_here.var_names)
    df['group']=adata_here.obs[grouping_variable]
    groups=list(set(df['group']))
    pseudobulk=df.groupby('group').sum()
    
    return(pseudobulk)
    
out1=create_pseudobulk(single2[:,common_genes],'gene.compact')
```
